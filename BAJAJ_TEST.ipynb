```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("/content/DataEngineeringQ2.json")
```

```python
df.head()
```

```python
df1 = df['patientDetails']
```

```python
df1.head()
```

```python
df2 = pd.DataFrame(df1.tolist())
```

```python
df2.head()
```

```python
m_fn = df2['firstName'].apply(lambda x: x== '' or x== None).sum()
total_fn  = len(df2['firstName'])
percentage_fn = round((m_fn/total_fn)*100,2)
print(percentage_fn)
```

```python
m_ln = df2['lastName'].apply(lambda x: x== '' or x== None).sum()
total_ln  = len(df2['lastName'])
percentage_ln = round((m_ln/total_ln)*100,2)
print(percentage_ln)
```

```python
m_db = df2['birthDate'].apply(lambda x: x== '' or x== None).sum()
total_db  = len(df2['birthDate'])
percentage_db = round((m_db/total_db)*100,2)
print(percentage_db)
```

```python
df2['gender'].fillna(df2['gender'].mode()[0], inplace = True)
```

```python
df2['gender']
```

```python
fc = df2['gender'].value_counts()['F']
```

```python
total = len(df2['gender'])
```

```python
percentage_f = round((fc/total)*100,2)
print(percentage_f)
```
